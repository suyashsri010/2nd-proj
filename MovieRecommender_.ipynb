Filtering the data based on our needs to recommend movies
i.e. which attributes are important to recommend movies

```python
#genres
#id
#keywords
#title
#overview
#cast
#crew

#re-initialising our original df with the new df containing attributes of our need
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.info()
```

```python
movies.isnull().sum() #finds if there is any empty space or no
```

```python
movies.dropna(inplace=True) #drops the rows carrying empty spaces
```

```python
movies.duplicated().sum() #finds duplicates
```

```python
movies.iloc[0].genres
```

```python
import ast #ast.literal_eval(  ) converts input string into a list
```

```python
def convert(obj): #function to filter the name from genres and keywords and remove unnecessary details
  L=[]
  for i in ast.literal_eval(obj): #converts string to list
    L.append(i['name'])
  return L
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj): #only takes 3 cast members
  L=[]
  counter=0
  for i in ast.literal_eval(obj): #converts string to list
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
def fetch_director(obj): #function to fetch director name from crew list
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break;
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head(1)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split()) #splits our sting with commas
# we do this because we have to create tags for the recommender system to use
movies.head(1)
```

```python
#now we remove space bw any two words without a comma
#we do this because our model can confuse bw two names
#example: sam bahadur and sam jones , as sam is both in names , so we combine
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies.head(1)
```

```python
#create a new column - named as tags
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
#create a new df
new_df=movies[['movie_id','title','tags']]
new_df.head(1)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df.head(1)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
#now we have a vector of all movies with corresponding values to the number of occurences of the top 5000 words
#to find out the top 5 most similar movie to a single movie, we may use euclidean distance bw vectors
#however it is not a good measure when dealing with too large of dimensions like 5000
#therefore better would be to use cosine distance bw the vectors that is the angle bw them
#distance is inverse to similarity,more distance less similar- less distance bw two vectors more similar
#we import cosine similarity from scikit learn

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors) #4806*4806 every movie with every movie
#diagnol will always be 1
```

define function which returns 5 similar movies to an input movie

```python
#find index of input movie
#find it in similarity matrix using similarity[index]
#sort according to similarity
#return second last to sixth last movie as last movie is the same movie with max similarity
def recommend(movie):
  movie_index = new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
  return
```
